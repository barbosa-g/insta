```python
import pandas as pd
pd.options.display.float_format='{:,.2f}'.format
url = "https://github.com/barbosa-g/insta/raw/main/08.%20Analisando%20o%20engajamento%20no%20Instagram.xlsx"
o_base = pd.read_excel(url)
```

```python
# Rename columns headers
base = o_base.rename(columns={
    'Tipo': 'Type',     'Data': 'Date',     'Curtidas': 'Likes',     'Comentários': 'Comments',
    'Pessoas': 'People',     'Campanhas': 'Campaign',    'Visualizações' : 'Views',    'Carrossel': 'Carousel',
    'Interacoes ': 'Interactions'})
```

```python
base = base.drop("Views",axis=1)
base.head()
```

```python
base.info()
```

```python
# convert to String
base['Type'] = base['Type'].astype(str)
base['Tags'] = base['Tags'].astype(str)
base['People'] = base['People'].astype(str)
base['Campaign'] = base['Campaign'].astype(str)
base['Carousel'] = base['Carousel'].astype(str)
```

```python
# assign results back to DataFrame (base)
base[['Type', 'Tags', 'People', 'Campaign', 'Carousel']] = base[['Type', 'Tags', 'People', 'Campaign', 'Carousel']].astype(str)
```

```python
# check nulls
base.Tags.isnull().value_counts()
```

```python
# 2 tags same ROW, Mean cannot be calculated properly
base.groupby('Tags')['Likes'].mean()
```

```python
# Split Tags
base.Tags  = base.Tags.str.split("/")
base.head()
```

```python
# 1 Tag each ROW - Using .explode saving in a new variable
# explode Tags - will help calc. Tags, but will not help with other columns
base_tag = base.explode('Tags')
base_tag.head(5)
```

```python
# Average/Mean - now its Correct
base_tag.groupby('Tags')['Likes'].mean()
```

```python
# Its ok to calculate other columns, since the variable base is not with 'explode'
base.groupby('People')['Likes'].mean()
```

```python
base.head()
```

```python
# order and sort
base_tag.groupby('Tags')[['Likes', 'Comments']].mean().sort_values('Likes',ascending=False)
```

```python
# checkin for null in Tags and attribute 'No Tag'
base_tag.loc[base_tag.Tags.isnull(),'Tags'] = 'No Tag'
```

```python
# Calculate Mean, Sort values again - with 'No Tag'
base_tag.groupby('Tags')[['Likes', 'Comments']].mean().sort_values('Likes',ascending=False)
```

```python
# if needed, attribute Null values
import numpy as np
base_tag.loc[base_tag.Tags == 'No Tag'] = np.nan
base_tag.groupby('Tags')[['Likes', 'Comments']].mean().sort_values('Likes',ascending=False)
```

```python
base_tag.groupby(['People','Tags'])[['Likes', 'Comments']].mean().sort_values('Likes',ascending=False)
```
